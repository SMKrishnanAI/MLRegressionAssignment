```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#loading data
dataframe=pd.read_csv("insurance_pre.csv")
```

```python
dataframe.head()
```

```python
dataframe.columns
```

```python
# Convert Categorical to Dummy Variables
dataframe_processed = pd.get_dummies(dataframe, columns=['sex', 'smoker'], drop_first=True)
```

```python
dataframe_processed.head()
```

```python
dataframe_processed.columns
```

```python
#independent=dataframe_processed[['age', 'bmi', 'children',  'sex_male', 'smoker_yes']]
```

```python
# 2. Define X and y which is independent and dependent
independent = dataframe_processed.drop('charges', axis=1) #X
independent.head()
```

```python
dependent=dataframe_processed[['charges']] #Y
dependent.head()
```

```python
#from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.3, random_state=0)
```

### from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)

```python
weight=regressor.coef_
weight
```

```python
bias=regressor.intercept_
bias
```

```python
y_predict=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
rscore=r2_score(y_predict,y_test)
rscore
```

```python
import pickle
pickle.dump(regressor,open("Wk3_Asgn_mliear_001.sav","wb"))
```

```python
# Deploying saved model 
loaded_model=pickle.load(open("Wk3_Asgn_mliear_001.sav","rb"))
```

```python
#'age', 'bmi', 'children',  'sex_male', 'smoker_yes
age_input=float(input("Age:"))
bmi_input=float(input("Bmi: "))
children_input=float(input("Number of Children: "))
sex_input=int(input("Male 1 or Female 0: "))
smoker_input=int(input("Smoker 1 or Non Smoker 0: "))
```

```python
result=loaded_model.predict([[age_input,bmi_input,children_input,sex_input,smoker_input]])
```
