```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#reading the data into dataframe
dataframe=pd.read_csv("insurance_pre.csv")
```

```python
dataframe.head()
```

```python
dataframe_processed=pd.get_dummies(dataframe, drop_first=True)
```

```python
dataframe_processed.head()
```

```python
dataframe_processed.columns
```

```python
independent=dataframe_processed[['age', 'bmi', 'children', 'sex_male', 'smoker_yes' ]]
independent.head()
```

```python
dependent=dataframe_processed[[ 'charges']]
dependent.head()
```

```python
# #split into training set and test set
# from sklearn.model_selection import train_test_split
# X_train,X_test,y_train,y_test= train_test_split(independent, dependent, test_size=1/3, random_state=0)
```

```python
# from sklearn.preprocessing import StandardScaler
# sc=StandardScaler()
# X_train=sc.fit_transform(X_train)
# X_test=sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
#from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
param_grid = { 'criterion' : ['squared_error','absolute_error','friedman_mse','poisson'],
                'max_features' : ['auto','sqrt','log2'],
                'n_estimators' : [10,100]
              }
```

```python
grid=GridSearchCV(RandomForestRegressor(), param_grid, refit=True, verbose = 3, n_jobs = -1)
```

```python
#grid.fit(X_train,y_train)
grid.fit(independent,dependent)
```

```python
#print best parameter after tuning
print(grid.best_params_)
```

```python
re=grid.cv_results_
```

```python
print("The  r_Score value for the best parameter {}".format(grid.best_params_))
```

```python
#re
table=pd.DataFrame.from_dict(re)
```
